```python
# FirstName_LastName_EDA.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets from data/ directory
customers_df = pd.read_csv('../data/Customers.csv')
products_df = pd.read_csv('../data/Products.csv')
transactions_df = pd.read_csv('../data/Transactions.csv')

# Inspect the datasets
print(customers_df.head())
print(products_df.head())
print(transactions_df.head())

# Check for missing values
print(customers_df.isnull().sum())
print(products_df.isnull().sum())
print(transactions_df.isnull().sum())

# EDA on Customers
# Distribution of customers by region
plt.figure(figsize=(8,6))
sns.countplot(data=customers_df, x='Region')
plt.title('Customer Distribution by Region')
plt.savefig('../reports/Customer_Distribution_by_Region.png')
plt.show()

# EDA on Products
# Top product categories by price
plt.figure(figsize=(8,6))
sns.boxplot(data=products_df, x='Category', y='Price')
plt.title('Price Distribution by Product Category')
plt.savefig('../reports/Price_Distribution_by_Product_Category.png')
plt.show()

# EDA on Transactions
# Total sales per product
sales_per_product = transactions_df.groupby('ProductID')['TotalValue'].sum().reset_index()
sales_per_product = sales_per_product.merge(products_df[['ProductID', 'ProductName']], on='ProductID')

plt.figure(figsize=(10,8))
sns.barplot(data=sales_per_product, x='TotalValue', y='ProductName')
plt.title('Total Sales per Product')
plt.savefig('../reports/Total_Sales_per_Product.png')
plt.show()

# Generate insights and save a PDF with these findings in the reports folder.
# You can use LaTeX, report libraries, or manual creation of PDF reports in your workflow.
```
